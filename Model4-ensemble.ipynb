```python
import pandas as pd
import numpy as np

#from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt

import os          
os.getcwd()

data = pd.read_csv('final_data.csv', engine='python')    # Supply the file name (path)
data.head(5)      
```

```python
cleaned_reviews = pd.read_csv('cleaned_reviews.csv', engine='python')
data['cleaned_reviews'] = cleaned_reviews
data.head(5)
```

```python
# Splitting the dataset into the Training set and Test set
X = data.cleaned_reviews
y = data.revenue_class

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#vect = CountVectorizer()
#X_train_dtm = vect.fit_transform(X_train)
#X_train_dtm.shape

# include 1-grams and 2-grams
vect = CountVectorizer()
X_train_dtm = vect.fit_transform(X_train)
X_train_dtm.shape

X_test_dtm = vect.transform(X_test)
X_test_dtm.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score
from sklearn import metrics

vect = CountVectorizer()

#Naive Bayes
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

# create a pipeline of vectorization and Naive Bayes
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(vect, nb)

# examine the pipeline steps
pipe.steps
```

```python
# cross-validate the entire pipeline
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X, y, cv=5, scoring='accuracy').mean()
```

```python
pipe.fit(X_train, y_train)
y_pred_class_pipe = pipe.predict(X_test)
metrics.confusion_matrix(y_test, y_pred_class_pipe)
```

```python
metrics.accuracy_score(y_test, y_pred_class_pipe)
```

```python
from sklearn.grid_search import RandomizedSearchCV
import scipy as sp
param_grid = {}
param_grid['countvectorizer__min_df'] = [1, 2, 3]
param_grid['countvectorizer__max_df'] = [1.0, 0.9, 0.8]
#param_grid['countvectorizer__ngram_range'] = [1, 2]
param_grid['multinomialnb__alpha'] = sp.stats.uniform(scale=1)
param_grid
```

```python
# set a random seed for sp.stats.uniform
np.random.seed(1)

from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score
from sklearn import metrics

vect = CountVectorizer()

#Naive Bayes
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

# create a pipeline of vectorization and Naive Bayes
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(vect, nb)

# examine the pipeline steps
pipe.steps

# additional parameters are n_iter (number of searches) and random_state
rand = RandomizedSearchCV(pipe, param_grid, cv=5, scoring='accuracy', n_iter=5, random_state=1)
```

```python
# time the randomized search
%time rand.fit(X_train, y_train)
```
